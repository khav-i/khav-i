# **Book trail**

В этой тетрадке я веду записи о своём читательском пути. На чужой взгляд он, конечно, может предстать каким угодно, но публикую я его только за тем, чтобы помочь тому, кого случайно чем-то мог увлечь, быстрее подобрать тон восприятия моей персоны.

Начну, пожалуй, с замечания (справедивого, полагаю, по отношению ко многим во многих аспектах их человеческого быта), что ничего из того, о чём я мыслю, вероятней всего, не обязано **мне** своим происхождением: я лишь плавильня мнений, идей и убеждений, давно вытоптанных на ближайшем круге блужданий мысли человека за всю историю существования последнего. Вся прелесть в том, наверное, чтобы быть чуточку другой плавильней смыслов — уникальных, если сильно повезёт, а если повезёт ещё сильнее — смыслов достойных. Если что-то такое вдруг могло стать целью вашей жизни, то о дороге к ней я вам ничего поведать, увы, не сумею.

Разумеется, *глубина личности* не измеряется в книгах, прочитанным не исчерпывается *описание души*, но подобный список всегда есть некоторым образом запечатлённый путь внимания, уделённого вещам. Насколько пристального — вопрос открытый.

Я отмечаю дату занесения прочитанной книги в список (д.з.с.), и по ним, где они есть, можно косвенно судить о том, что меня могло в то время занимать. Хотя, надо признаться, я нередко читаю то, что было отложено уже давно, просто потому что — а чего лежит-то книга пылится?! Привычка складирования книг может отнять у вас жизнь.

В примечаниях будут мелькать звёзды — они отмечают значимость того или иного произведения для развития моей мысли, формирования ключевых убеждений.

Иногда способ группировки произведений в одну запись может показаться странным: да, единого принципа у меня не было — объединял по переплёту, по спонтанному очущению меры.

Список пополняется сверху, поэтому чем дальше вниз, тем сильнее в прошлое мы уходим.
